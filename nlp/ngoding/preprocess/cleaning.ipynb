```python
import pandas as pd
import re
import numpy as np
from textblob import TextBlob
```

```python
#merge df
df1 = pd.read_csv('https://raw.githubusercontent.com/Zaniiiii/AdvanceStudyAICenter/main/nlp/data/ganjar_1.csv', sep=',')
df2 = pd.read_csv('https://raw.githubusercontent.com/Zaniiiii/AdvanceStudyAICenter/main/nlp/data/ganjar_2.csv', sep=',')

df = pd.concat([df1,df2], axis = 0)
df
```

```python
# reset index concat result
df = df.reset_index(drop=True)

# drop columns except text
df_clean = df[['full_text']]
df_clean
```

```python
def remove_username(tweet_text):
    return re.sub(r'@[\w]+', '', tweet_text)

df_clean['full_text'] = df_clean['full_text'].apply(remove_username)
df_clean
```

```python
df_clean.info()
```

```python
df_clean['full_text'] = df_clean['full_text'].astype(pd.StringDtype())
df_clean.info()
```

```python
for i in range(len(df_clean)):
  tweet = df_clean["full_text"][i]
  df_clean["full_text"][i] = ' '.join(re.sub("(@[A-Za-z0-9]+)|([^0-9A-Za-z \t])|(\w+:\/\/\S+)"," ",tweet).split())
```

```python
df_clean
```

```python
duplicate = df_clean[df_clean.duplicated("full_text")]

print("Duplicate Rows :")

# Print the resultant Dataframe
duplicate
```

```python
df_clean.drop_duplicates(subset="full_text", keep=False, inplace=True)
df_clean = df_clean.reset_index(drop=True)

df_clean
```
